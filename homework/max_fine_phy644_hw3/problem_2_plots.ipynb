We are making a plot of: 

$$
    \boxed{r(\eta) = \frac{GM}{2|\epsilon|}(1-\cos\eta)} $$
   $$ \boxed{t(\eta) = \frac{GM}{\sqrt{8|\epsilon|^3}} (\eta - \sin \eta)}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants (set GM=1, |epsilon|=1 for simplicity; scale as needed)
GM = 1.0
eps = 1.0

# Parameter eta, only one cycle (0 to 2)
eta = np.linspace(0, 2*np.pi, 500)

# Equations
r = (GM / (2*eps)) * (1 - np.cos(eta))
t = (GM / np.sqrt(8*eps**3)) * (eta - np.sin(eta))

# Find maximum extent
imax = np.argmax(r)
t_max, r_max = t[imax], r[imax]

# Plot
plt.figure(figsize=(8,6))
plt.plot(t, r, color='purple')
plt.axvline(t_max, color='gray', linestyle='--', label="Max extent")

# Labels and annotations
plt.xlabel(r"$t(\eta)$", size=15)
plt.ylabel(r"$r(\eta)/r_t$", size=15)
plt.title("Parametric Cycloidal Curve\n", size=15)

# Annotate max extent
plt.text(t_max, r_max*1.05, f"max at =", ha='center', va='bottom', fontsize=12)

# Annotate when it returns to zero (=2)
plt.text(t[-1], r[-1]+0.05, r"returns at $\eta=2\pi$   ", 
         ha='right', va='bottom', fontsize=12, color="k")

plt.legend()
plt.show()
```
