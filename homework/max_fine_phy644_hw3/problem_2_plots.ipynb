```python
# Creating a plot of r(t) for the bound (<0) parametric solution.
import numpy as np
import matplotlib.pyplot as plt

# Use nondimensionalized variables:
# r/A = 1 - cos(eta), t/B = eta - sin(eta)
eta = np.linspace(1e-6, 2*np.pi - 1e-6, 1000)  # avoid exact 0 and 2pi to avoid numerical issues
r_over_A = 1 - np.cos(eta)
t_over_B = eta - np.sin(eta)

# now pick physical constants for an illustrative curve
GM = 1.0
eps_abs = 0.1  # ||
A = GM / (2.0 * eps_abs)          # A = GM/(2||)
B = GM / (2.0 * eps_abs)**1.5    # B = GM/(2||)^{3/2}

t = B * t_over_B
r = A * r_over_A

plt.figure(figsize=(7,4.5))
plt.plot(t, r, color='green')
plt.xlabel(r'$t$', size=15)
plt.ylabel(r'$r(t)$', size=15)
plt.title(r'Radial cycloid solution $r(t)$ for bound case ($\varepsilon<0$)', size=15)
plt.axvline(B*np.pi, linestyle='--', linewidth=0.8, color='purple', lw=2)  # turnaround time at eta = pi
plt.text(B*np.pi, A*1.02, 'turnaround\n(=)', ha='center', va='bottom', fontsize=15)
plt.tight_layout()
plt.show()

# Also plot nondimensional universal curve r/A vs t/B
plt.figure(figsize=(7,4.5))
plt.plot(t_over_B, r_over_A, color='green')
plt.xlabel(r'$\;t/B$', size=15)
plt.ylabel(r'$r/A$', size=15)
plt.title(r'Universal nondimensional curve: $r/A = 1-\cos\eta,\; t/B=\eta-\sin\eta$', size=15)
plt.axvline(np.pi, linestyle='--', color='purple', lw=2)
plt.text(np.pi, 1.02, 'turnaround\n(=)', ha='center', va='bottom', fontsize=15)
plt.tight_layout()
plt.show()

```
